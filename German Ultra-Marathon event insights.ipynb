```python
# Some values in the Athlete average speed are calculated wrong or are in the form of time which appears to be copied from the Athlete Performance column.
# Calculating the Athlete average speed using the Event Distance/Length and Athlete performance columns
```

```python
# Values in the Event distance/length is km as well as miles and the Athlete average speed is in km/hr. Hence unit conversion is required.
```

```python
df2['Event distance/length'].str.strip()
#distance, unit = df2['Event distance/length'].str[:-2], df2['Event distance/length'].str[-2:]
```

```python
# Convert distance to kilometers
def convert_to_km(distance_str):
    distance, unit = float(distance_str[:-2]), distance_str[-2:]
    if unit == 'mi':
        return distance * 1.60934
        # converting miles to kilometers
    return distance

# Applying the conversion function to the 'Event distance/length' column
distance_km = df2['Event distance/length'].apply(convert_to_km)

# Converting the Athlete performance values to timedelta (hours)
athlete_time = pd.to_timedelta(df2['Athlete performance'])
#print(athlete_time)

# Calculating the speed in km/h and amending the 'Athlete average speed' column
df2['Athlete average speed'] = distance_km / (athlete_time.dt.total_seconds() / 3600)
```

```python
# Effect of Seasons on Performance
df4 = df3.copy()
df4.head()
```

```python
# Converting the Event Date column data type
# Dealing with the 'event_date' column values that contains a hyphen (to indicate a date range)
df4['event_date'] = df4['event_date'].apply(lambda x: x.split('-')[1] if '-' in x else x)
df4['event_date'] = pd.to_datetime(df4['event_date'], format='%d.%m.%Y')
df4.dtypes
```

```python
df4.head()
```

```python
df4['season'] = df4['event_date'].dt.month.apply(
    lambda x: 'winter' if x == 12 or x <= 2 else
    'spring' if x == 3 or x == 4 or x == 5 else
    'summer' if x == 6 or x == 7 or x == 8 else
    'fall'  # from September (9) to November (11)
)
```

```python
df4.season.value_counts()
```

```python
# Average speed by season
season_performance = df4.groupby('season')['athlete_avg_speed'].agg(['mean', 'count']).sort_values(by='mean',ascending=False)
season_performance
```

```python
# Pie Chart
plt.pie(season_performance['count'], labels=season_performance.index, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.show()
```

```python
# Average speed by season and gender
df4.groupby(['season','athlete_gender']).agg({'athlete_gender': 'count', 'athlete_avg_speed': ['mean', 'count']})
```

```python
sns.relplot(season_performance, x='season', y='mean', kind = 'line')
plt.title('Average Speed by Season')
plt.show()
```
